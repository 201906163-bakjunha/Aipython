```python
import numpy as np
from sklearn import linear_model # from sklearn.linear_model.LinearRegression
import matplotlib.pyplot as plt
regr = linear_model.LinearRegression() #reg = LinearRegression()
X = [[164], [179], [162], [170]]
y = [53, 63, 55, 59]
regr.fit(X, y)
plt.scatter(X, y)
y_pred = regr.predict(X)
plt.plot(X, y_pred, color = 'blue', linewidth = 3)
plt.show()
```
