```python
data={'도시': ['서울', '부산', '인천', '대구'],
      '인구 수': [904312, 3348737, 289451, 2466042]}
df=pd.DataFrame(data)
df1=df.sort_values('인구 수', ascending=False)
plt.bar(df1['도시'], df1['인구 수'])
plt.title('도시별 인구 수')
plt.xlabel('도시')
plt.ylabel('인구 수')
```

```python
import seaborn as sns
iris_data=pd.read_csv('/content/drive/MyDrive/실습/11주차/iris.csv')
sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data= iris_data, legend=False)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```
